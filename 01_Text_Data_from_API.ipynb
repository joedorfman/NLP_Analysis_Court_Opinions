```python
header = {'Authorization': 'Token '}
```

```python
def get_text(id):
    URL_base2="https://www.courtlistener.com/api/rest/v3/opinions/"
    URL=URL_base2+str(id)
    print (id)
    print (URL)
    response = requests.get(URL, headers=header )
    try:
        response_dict = response.json()
        response_html = response_dict['html']
        response_soup = BeautifulSoup(response_html, 'lxml')
        response_text = response_soup.get_text()
        return response_text
    except:
        print ("Error!")
        return np.nan
    

def get_text2(id):
    URL_base2="https://www.courtlistener.com/api/rest/v3/opinions/"
    URL=URL_base2+str(id)
    print (id)
    print (URL)
    response = requests.get(URL, headers=header )
    try:
        response_dict = response.json()
        response_text = response_dict['plain_text']
        if response_text=="":
            response_html = response_dict['html']
            response_soup = BeautifulSoup(response_html, 'lxml')
            response_text = response_soup.get_text()
        return response_text
    except:
        print ("Error!")
        return np.nan
```

```python
circuit['text']=circuit['id'].map(get_text2)
```
